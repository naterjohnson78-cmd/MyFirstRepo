```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests
from bs4 import BeautifulSoup

response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/employees_3.html').text
```

```python
soup = BeautifulSoup(response, 'html.parser')
print(soup)
```

```python
soup.find('table')
```

```python
soup.find('table').find_all('tr', class_='employee')[0].find_all('td')[1].text
```

```python
for employee in soup.find('table').find_all('tr', class_='employee'):
    print(employee.find_all('td'))
```

```python
for employee in soup.find('table').find_all('tr', class_='employee'):
    print(employee.find_all('td')[1].text)
```

```python
name = []

for employee in soup.find('table').find_all('tr', class_='employee'):
    name.append(employee.find_all('td')[1].text)

print(name)
```

```python
age = []

for employee in soup.find('table').find_all('tr', class_='employee'):
    name.append(employee.find_all('td')[1].text)
    age.append(employee.find_all('td')[2].text)

print(age)
```

```python
average_age = sum(int(age) for age in age) / len(age)
print(average_age)
```

```python
ids = []
names = []
ages = []

for employee in soup.find('table').find_all('tr', class_='employee'):
    tds = employee.find_all('td')
    ids.append(tds[0].text)
    names.append(tds[1].text)
    ages.append(tds[2].text)

df = pd.DataFrame({
    'id': ids,
    'name': names,
    'age': ages
})

df.head()
df.head()
```

```python
df.info()
```

```python
# convert lists in-place
age[:] = [int(x) for x in age]
ages[:] = [int(x) for x in ages]

# convert dataframe column to integers
df['age'] = df['age'].astype(int)

# quick check
print(age)
print(df['age'].dtype)
```
