```python
import pandas as pd

import requests
from bs4 import BeautifulSoup
response = requests.get('https://btech-data-analytics.github.io/bridgerland-technical-college/bookstore.html').text
soup = BeautifulSoup(response, 'html.parser')
```

```python
books = soup.find_all('tr', class_='book')

titles = [book.find_all('td')[1].text for book in books]
authors = [book.find_all('td')[2].text for book in books]
prices = [float(book.find_all('td')[6].text.replace('$', '')) for book in books]
pages = [int(book.find_all('td')[4].text) for book in books]

df = pd.DataFrame({
    'title': titles,
    'author': authors,
    'price': prices,
    'pages': pages
})

df.head()
```

```python
df['author'].value_counts().head(3)
```

```python
topics = [book.find_all('td')[5].text for book in books]
topic_counts = pd.Series(topics).value_counts()
print(topic_counts)
print("Most popular topic:", topic_counts.idxmax())
```

```python
max_average_price = df.groupby(topics)['price'].mean().sort_values(ascending=False)
max_average_price.head(3)
```

```python
most_pages_average = df.groupby(topics)['pages'].mean().sort_values(ascending=False)
most_pages_average.head(3)
```
