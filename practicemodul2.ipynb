```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

df = pd.read_csv('companies.csv')
```

```python
df.head()
```

```python
df.groupby('company_name').mean(numeric_only=True)
```

```python
plt.bar(df['company_name'], df['estimated_revenues'])
plt.xlabel('Company Name')
plt.ylabel('Revenue')
plt.title('Revenue by Company')
plt.xticks(rotation=90)
plt.show()
plt.xlabel('Company Name')
plt.ylabel('Revenue')
plt.title('Revenue by Company')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='state', data=df)
plt.show()
```

```python
totaljobsdf = df.groupby('Industry', as_index=False).agg({'job_openings': 'sum'})
totaljobsdf
```

```python
sns.barplot(y='Industry', x='job_openings', data=totaljobsdf)
plt.xlabel('Industry')
```

```python
df.groupby('founded').mean(numeric_only=True).agg({'growth_percentage': 'mean'})
```

```python
sns.lineplot(x='founded', y='growth_percentage', data=df)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='employees', y='estimated_revenues', data=df)
plt.xlabel('Number of Employees')
plt.ylabel('Estimated Revenues')
plt.title('Employees vs Estimated Revenues')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='founded', bins=df['founded'].nunique(), kde=False)
plt.xlabel('Founded Year')
plt.ylabel('Number of Companies')
plt.title('Number of Companies Founded per Year')
plt.xticks(rotation=45)
plt.show()
```
